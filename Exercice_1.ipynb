```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
archivo = '/content/Iris.csv'
pf = pd.read_csv(archivo, sep=",", encoding='Latin-1')
```

```python
pf.head()
```

```python
descriptive_stats = pf.describe()
print("Descriptive statistics:")
print(descriptive_stats)
```

```python
# Size of the dataset(observations and number of variables)
num_observations = pf.shape[0]
num_variables = pf.shape[1]
print("Number of observations:", num_observations)
print("Number of variables:", num_variables)
```

```python
# Structure of dataset
print("\nStructure of dataset:")
print(pf.info())
```

```python
# Number of missing values now this
missing_values = pf.isnull().sum().sum()
print("\nNumber of missing values:")
print(missing_values)
```

```python
# Number of missing values per variable
print("\nNumber of missing values per variable:")
print(pf.isnull().sum())
```

```python
# Histograms
pf.hist(figsize=(10,6))
plt.suptitle('Histogramas of variables', y=1.05)
plt.tight_layout()
plt.show()
```

```python
# Boxplot and whisker plot
plt.figure(figsize=(10,6))
pf.boxplot()
plt.title('Box and Whisker Plots of Variables in the Iris Dataset')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histograms of each variable
num_columns = len(pf.columns)
num_rows = (num_columns + 1) // 2  # Calculates number of rows needed

plt.figure(figsize=(12, 8))
for i, column in enumerate(pf.columns):
    plt.subplot(num_rows, 2, i+1)
    plt.hist(pf[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Select variables to compare
variables_to_compare = ['SepalLengthCm', 'PetalLengthCm']

# Create boxplots to show the distribution of the selected variables
sns.boxplot(data=pf[variables_to_compare])
plt.title('Boxplots for Selected Variables')
plt.xlabel('Variable')
plt.ylabel('Value (cm)')
plt.show()

```

```python
# Scatter plot
plt.figure(figsize=(10, 6))

# SepalLengthCm vs SepalWidthCm
plt.scatter(pf['SepalLengthCm'], pf['SepalWidthCm'], label='Sepal', color='blue')

# PetalLengthCm vs PetalWidthCm
plt.scatter(pf['PetalLengthCm'], pf['PetalWidthCm'], label='Petal', color='red')

plt.xlabel('Length (cm)')
plt.ylabel('Width (cm)')
plt.title('Scatter Plot of Iris Dataset')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#The relationship between sepal length and petal length. It selects these two variables from the DataFrame
variables_a_comparar = ['SepalLengthCm', 'PetalLengthCm']

#  Dependencies between the variables
sns.pairplot(pf[variables_a_comparar])
plt.suptitle('Diagramas de dispersión para variables seleccionadas', y=1.02)
plt.show()
```

```python
# The mean of each feature per species
mean_by_species = pf.groupby('Species').mean()

# Bar chart
mean_by_species.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Species')
plt.ylabel('Average Measurement (cm)')
plt.title('Average Measurements by Species')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Feature')
plt.show()
```

```python
# Calculate the count of each species
species_counts = pf['Species'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Species in Iris Dataset')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
